# Proyecto Data Analysis
## Data Science Bootcamp
En este proyecto vas a aplicar todo lo aprendido en el Ramp Up y el mdulo de Data Analysis del bootcamp. Tendrs que realizar un anlisis exploratorio de datos y presentar los resultados. Ten en cuenta lo siguiente:

* **Miembros**: El proyecto es individual
* **Datos**: Tienes que conseguir los datos por tu cuenta (ver abajo)
* **Fechas**: Consultar calendario de Google Classroom


## Objetivo
Lo que queremos mediante este tipo de anlisis es resolver un problema real apoyndonos en datos, por lo que tendrs que plantearte primero qu es lo que quieres resolver, y a partir de ahi ir trabajando con los datos para acabar llegando a una conclusin. Plantea tus hiptesis antes de realizar el anlisis exploratorio y despus verifica si se han cumplido, o si has descubierto mediante los datos que lo que ocurra era justamente lo opuesto a lo que pensabas.


## Evaluacin

## Presentacin
Puesto que en la vida real la mayora de presentaciones se realizan a un pblico no tcnico, sta no va a ser menos. Si tienes que explicar algo tcnico, cuntalo, pero para alguien que no sabe lo que significa. Por ejemplo, cmo contaras qu es una matriz de correlacin? No queremos ver cdigo, ni cmo estrucutraste el proyecto en carpetas, ni cmo de difcil ha sido limpiarlo. Al cliente/tu jefe le da igual :). Explicar cosas tcnicas lleva tiempo por lo que piensa muy bien en lo que vas a decir y en cmo lo sintetizars.

Tendrs que exponer bien tu caso de uso (30% del tiempo), explicar qu es lo que has hecho para resolverlo (50%), y acabar la presentacin con unas conclusiones y oportunidades que nos brinda el proyecto (20%), para cerrar el plantemiento inicial del caso de uso. Los % son orientativos... 

**Formatos de la presentacin?** Nada de Notebooks, ni informes en pdf o HTML. Formato presentacin en PowerPoint, Prezi o la plataforma que mejor conozcas, siendo perfectamente combinable con un dashboard desarrollado mediante alguna de las herramientas aprendidas durante el curso (Power BI, Dash, Plotly).

**Consejos**: Si usas un PowerPoint... ser tu apoyo a la hora de presentar, aqu el protagonista es tu discurso. Por favor, que la presentacin no sea un conjunto de puntos inconexos, intenta llevar un hilo, como si de una historia se tratase. Vndete, destaca las fortalezas de tu trabajo sobre las debilidades. Interacta con el pblico. Si vas a contar algo de negocio, puedes traducirlo a nmeros? Qu impacto tendr tu analtica?. Muestra algunos datos concretos (estadsticos simples) para justificar tu business case o para cerrar tu presentacin. Si has elegido una temtica que est acorde con tus gustos/background, aprovecha al principio para contarlo y empatizar mucho ms con el pblico. Si vas a hacer una demo, grbala por si te falla algo en la presentacin (conexin a Internet). Preprate las respuestas a algunas posibles preguntas en un anexo de la presentacin.

**NOTA**: existen muchas plantillas gratuitas en la web para hacer presentaciones. [Aqu tienes unas cuantas](https://www.free-powerpoint-templates-design.com/free-powerpoint-templates-design/).

## Memoria
Simplemente es un documento que complemente a la presentacin. Nada serio. Ya que en la presentacin no se va a ver reflejado todo el esfuerzo dedicado al proyecto, utiliza este documento para describir todo tu trabajo, qu tcnicas has utilizado, con qu dificultades te has encontrado. No ms de 5 hojas. Recuerda documentar bien las fuentes de datos utilizadas en la memoria.


## Pasos del proyecto

1. Elegir temtica
    1. **Elegir una temtica y buscar datos en funcin de la misma**. Es algo ms complejo que la segunda opcin ya que puede ocurrir que no encuentres datos que te sirvan para tu propsito, o que tengas que conseguirlos mediante APIs, scrapping, descarga de BBDD.... No obstante, esta opcin es la que ms se asemejara a un caso real, ya que una gran problemtica en este tipo de proyectos es encontrar datos.

    2. **Directamente ir a pginas de datos abiertos y escoger un dataset de tu inters**. Si no tienes clara la temtica puedes buscar en alguno de los enlaces de datos (recomendable Kaggle) y escoger un tema que te motive.

2. Obtencin de datos. Puedes llevar a cabo el proyecto con estos datos?
3. Define tu hiptesis. Qu piensas que puedes obtener de estos datos? Qu vas a poder resolver? Cmo lo vas a llevar a cabo?
4. Preprocesado: obtn todos los datos de las diferentes fuentes que hayas utilizado, jntalos y ponlos en un formato entendible.
5. Limpia los datos: duplicados, missings, columnas intiles...
6. Exploratorio: obtn todos los estadsticos y grficos que necesites para entender bien tu dataset.
7. Concluye con tu anlisis si estabas o no en lo cierto acerca de tu planteamiento y tu hiptesis.
8. (Bonus track). Crea un dashboard para representar tu anlisis exploratorio.
9. Documenta tu proyecto y sbelo a GitHub.

Intenta escoger un dataset que te pueda servir despus para realizar predicciones sobre alguna de tus variables de inters. Este punto no es obligatorio, pero si recomendable.
