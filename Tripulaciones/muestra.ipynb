```python
import json
from sqlalchemy import create_engine
import pandas as pd

# Leemos json con las credenciales
with open('settings_sql.json', "r") as json_file_readed:
        json_readed = json.load(json_file_readed)

# Asignamos valores
IP_DNS = json_readed["IP_DNS"]
PORT = json_readed["PORT"]
USER = json_readed["USER"]
PASSWORD = json_readed["PASSWORD"]
BD_NAME = json_readed["BD_NAME"]

SQL_ALCHEMY = 'mysql+pymysql://' + USER + ':' + PASSWORD + '@' + IP_DNS + ':' + str(PORT) + '/' + BD_NAME

# Creamos conexión
engine = create_engine(SQL_ALCHEMY)

```

```python
event_df = pd.read_excel('0_all_events.xlsx')
event_df.drop(['Unnamed: 0', 'id_evento'], axis=1, inplace=True)
event_df
```

```python
event_df.columns
```

```python
event_df.rename(columns={' accesibilidad': 'accesibilidad', 'Ciudad':'ciudad'}, inplace=True)
```

```python
event_df.accesibilidad = event_df.accesibilidad.apply(lambda x: x.replace('\\xF0\\x9F\\x93\\x9D', ''))
```

```python
event_df.to_sql('evento', engine, index=False, if_exists='append')
```
