### Step 1. Import the necessary libraries

```python

# remember to %matplotlib inline
%matplotlib inline
from sklearn.datasets import load_iris, load_boston
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# set seaborn style to white
sns.set_style("white")
```

### Step 2. Import the dataset of `chipotle` and assign it to a variable called chipo.

```python
address = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(address, sep='\t')
```

### Step 3. See the first 10 entries

```python
chipo.head(10)
```

### Step 4. Create a histogram of the top 5 items bought

```python
top5 = chipo.groupby('item_name').sum().sort_values('quantity', ascending=False).head(5)
top5['Item'] = top5.index
sns.catplot(data=top5.sort_values('quantity'), x='Item', y='quantity', kind='bar', color='royalblue').set_ylabels('Number of Times Ordered').set_xticklabels(rotation=90)
```

### Step 5. Create a scatterplot with the number of items orderered per order price

Make sure you get the same labels and title
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
orders = chipo.groupby('order_id').sum()
```

```python
sns.scatterplot(x=orders.item_price, y=orders.quantity, color='green', linewidth=0).set(xlabel='Items Ordered', ylabel='Order Price', title='Number of items ordered per order price')
```
