```python
import pandas as pd
import requests
import time

key_2 = 'a855df40'

def omdb_to_csv(ruta, row_start, row_num=999, key='7b2c6fff'):

    movies = pd.read_csv(ruta)
    data = []
    count = 1

    for titleid in movies.loc[row_start:(row_start+row_num), 'titleId']:
        omdb_url = f'http://www.omdbapi.com/?i={titleid}&apikey={key}'
        r = requests.get(omdb_url, params={'plot':'full'}).json()
        data.append(r)
        print(count, 'Response:', r['Response'])
        count += 1

    df = pd.DataFrame(data)
    df.to_csv(f'Data_OMDb_{row_start}.csv')
    return df

omdb_to_csv('data/IMDb_Clean_1.csv', 4000, key=key_2)
```
