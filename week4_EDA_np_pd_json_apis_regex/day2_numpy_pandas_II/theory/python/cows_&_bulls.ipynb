
- [Cows_and_Bulls]

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python
import random

def is_cow(digit, entrada):
    lista_c = ['c' if v == digit[i] else v for i, v in enumerate(entrada)]
    cows = sum([1 for i, v in enumerate(entrada) if v == digit[i]])
    return cows, lista_c


def is_bull(digit, entrada):
    bulls = 0
    l = list(digit)
    for i, v in enumerate(entrada):
        if v in digit and v != digit[i] and v in l:
            bulls += 1
            l.remove(v)
    return bulls
```

```python
cows, l = is_cow('1234', '1333')
print(l)
```

```python
def find_cow_bull(digit, entrada):
    cows, posibles = is_cow(digit, entrada)
    bulls = 0
    for i, v in enumerate(entrada):
        if v in posibles and posibles[i] == digit[i]:
            bulls += 1
    return cows, bulls

print(find_cow_bull('1350', '1330'))
```

```python
import random 

# Un número es generado de manera aleatoria
generated_num = str(random.randint(1000,9999))
generated_num = '1333'

def is_cow(digit, entrada):
    lista_c = [v for i, v in enumerate(entrada) if v == digit[i]]
    cows = sum([1 for i, v in enumerate(entrada) if v == digit[i]])
    return cows, lista_c

def find_cow_bull(digit, entrada):
    cows, eliminados = is_cow(digit, entrada)
    bulls = 0
    for i, v in enumerate(entrada):
        if v not in eliminados and v in digit and v != digit[i]:
            bulls += 1
    return cows, bulls

while True:
    guess_str = input('Introduzca numero:')
    if guess_str.isnumeric() and len(guess_str) == 4:
        cows, bulls = find_cow_bull(generated_num, guess_str)
        if cows == 4:
            print(f'Has ganado. El número era {generated_num}')
            break
        else:
            print(guess_str)
            print(f'Cows: {cows} | Bulls: {bulls}')
    elif guess_str == 'STOP':
        break
    else:
        print(f'{guess_str} no es un numero válido')
```
